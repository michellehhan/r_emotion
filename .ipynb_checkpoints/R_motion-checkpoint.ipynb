```python

# Bin Q39 into 3 levels for cross-tab: Low (-2 to -1), Neutral (0), High (1 to 2)
def bin_consistency(val):
    if pd.isna(val):
        return 'Unknown'
    elif val <= -1:
        return 'Low'
    elif val == 0:
        return 'Neutral'
    else:
        return 'High'

df_clean['Journaling_Consistency_Level'] = df_clean['Q39'].apply(bin_consistency)

# Now explode Q77 and count frequencies by consistency level
q77_exploded = df_clean[['Journaling_Consistency_Level', 'Q77']].dropna()
q77_exploded = q77_exploded.assign(Q77=q77_exploded['Q77'].astype(str).str.split(',')).explode('Q77')
q77_exploded['Q77'] = q77_exploded['Q77'].str.strip()

# Create cross-tab
cross_tab = pd.crosstab(q77_exploded['Q77'], q77_exploded['Journaling_Consistency_Level'])

# Normalize by row to get proportions
cross_tab_norm = cross_tab.div(cross_tab.sum(axis=1), axis=0).round(2)

# Prepare the data for seaborn: melt the crosstab into long format
df_plot = cross_tab_norm.reset_index().melt(id_vars='Q77', var_name='Consistency Level', value_name='Proportion')

# Sort features by total proportion (optional but helps with clarity)
feature_order = df_plot.groupby('Q77')['Proportion'].sum().sort_values(ascending=False).index.tolist()

# Define colors
color_map = {
    'High': '#2ca02c',  
    'Neutral': '#1f77b4',   
    'Low': '#ff7f0e',       
    'Unknown': '#7f7f7f'  
}

plt.figure(figsize=(16, 8))
sns.barplot(
    data=df_plot,
    x='Q77',
    y='Proportion',
    hue='Consistency Level',
    order=feature_order,
    hue_order=['Unknown', 'Low', 'High','Neutral'],
    palette=color_map,
    dodge=True
)

plt.title("Habit Feature Preferences (Q77) by Journaling Consistency Level (Q39)", fontsize=16)
plt.xlabel("Habit-Building Feature", fontsize=12)
plt.ylabel("Proportion of Users", fontsize=12)
plt.xticks(rotation=35, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(title="Consistency Level", title_fontsize=11, fontsize=10, bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import re

multi_cols = {
    'Q63': "Journal Structure Preference (Q63)",
    'Character Dev': "User GoalsThrough Journaling",
    'Q64': "Preferred Prompt Type (Q64)"
}

q63_options = [
    "Free-form writing",
    "Bullet journaling style (lists, tasks)",
    "Structured templates (mood tracker, gratitude list)",
    "Visual/Emoji journaling",
    "Narrative timeline (events described as a flowing story)",
    "Mind map/branching thoughts",
    "Voice-to-text journaling",
    "Mood-based layouts",
    "Photo-based journaling",
    "Other (please specify)"
]

q64_options = [
    "Free, open writing",
    "Light reflective prompts",
    "Structured deep reflection",
    "A mix depending on my mood"
]

freq_data = {}

for col, label in multi_cols.items():
    if col == 'Q63':
        q63_raw = df_clean[col].dropna().astype(str).str.lower()
        q63_counts = pd.Series({
            opt: q63_raw.str.contains(re.escape(opt.lower())).sum()
            for opt in q63_options
        }).sort_values(ascending=True)
        freq_data[label] = q63_counts
    elif col == 'Q64':
        q64_raw = df_clean[col].dropna().astype(str).str.lower()
        q64_counts = pd.Series({
            opt: q64_raw.str.contains(re.escape(opt.lower())).sum()
            for opt in q64_options
        }).sort_values(ascending=True)
        freq_data[label] = q64_counts
    else:
        col_cleaned = df_clean[col].dropna().astype(str).str.split(',').explode().str.strip()
        counts = col_cleaned.value_counts().sort_values(ascending=True)
        freq_data[label] = counts

# Horizontal bar chart
fig, axes = plt.subplots(len(freq_data), 1, figsize=(11, 5 * len(freq_data)))

if len(freq_data) == 1:
    axes = [axes]

for ax, (label, counts) in zip(axes, freq_data.items()):
    bars = ax.barh(counts.index, counts.values, color="#55a868", height=0.8)
    ax.set_title(label, fontsize=16)
    ax.set_xlabel("Count", fontsize=16)
    ax.set_xlim(0, max(counts.values) * 1.2)
    ax.tick_params(axis='y', labelsize=11)
    for bar in bars:
        ax.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
                int(bar.get_width()), va='center', fontsize=15)

plt.tight_layout()
plt.show()
```

# AI Preferences: What is its role?

```python
# Step 1: Bin Q71 (AI comfort) into 3 groups based on Likert score
def bin_ai_comfort(score):
    if score <= -1:
        return 'Uncomfortable'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Comfortable'

df_clean['AI_Comfort_Level'] = df_clean['Q71'].apply(bin_ai_comfort)

# Step 2: Explode Q23 (AI features selected)
q23_exploded = df_clean[['AI_Comfort_Level', 'Q23']].dropna()
q23_exploded = q23_exploded.assign(Q23=q23_exploded['Q23'].astype(str).str.split(',')).explode('Q23')
q23_exploded['Q23'] = q23_exploded['Q23'].str.strip()

# Step 3: Cross-tab and normalize
ai_cross_tab = pd.crosstab(q23_exploded['Q23'], q23_exploded['AI_Comfort_Level'])
ai_cross_tab_norm = ai_cross_tab.div(ai_cross_tab.sum(axis=1), axis=0).round(2)

# Step 4: Melt for plotting
df_ai_plot = ai_cross_tab_norm.reset_index().melt(id_vars='Q23', var_name='AI Comfort Level', value_name='Proportion')

# Step 5: Define colors
comfort_palette = {
    'Comfortable': '#2ca02c',
    'Neutral': '#1f77b4',
    'Uncomfortable': '#ff7f0e'
}

# Step 6: Plot
plt.figure(figsize=(14, 8))
sns.barplot(
    data=df_ai_plot,
    x='Q23',
    y='Proportion',
    hue='AI Comfort Level',
    order=df_ai_plot['Q23'].value_counts().index,
    palette=comfort_palette,
    hue_order=['Uncomfortable', 'Comfortable', 'Neutral']
)

plt.title("Preferred AI Features (Q23) by AI Comfort Level (Q71)", fontsize=16)
plt.xlabel("AI-Powered Feature", fontsize=12)
plt.ylabel("Proportion of Users", fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Comfort Level", title_fontsize=11, fontsize=10, bbox_to_anchor=(1.01, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
q101_cols = {
    'Q101_1': 'Sadness',
    'Q101_2': 'Anxiety/Fear',
    'Q101_3': 'Joy',
    'Q101_4': 'Inspiration',
    'Q101_5': 'Anger'
}

# Select and reshape
q101_long = df_clean[q101_cols.keys()].rename(columns=q101_cols)
q101_long = q101_long.melt(var_name='Emotion', value_name='Preference')
q101_long['Preference'] = q101_long['Preference'].astype(str).str.strip()

# Count combinations
q101_counts = q101_long.groupby(['Emotion', 'Preference']).size().reset_index(name='Count')

order = [
    'Mostly freedom',
    'Some freedom',
    'No preference',
    'Some guidance',
    'Mostly guidance'
]

palette = {
    'Mostly freedom': '#9ecae1',
    'Some freedom': '#6baed6',   
    'No preference': '#fdd0a2',
    'Some guidance': '#fc9272',      
    'Mostly guidance': '#de2d26'     
}

# Step 4: Plot grouped bar chart
plt.figure(figsize=(12, 5))
sns.barplot(data=q101_counts, x='Emotion', y='Count', hue='Preference', hue_order=order, palette=palette)
plt.title("When journaling emotions (Q101), is more guidance or freedom preferred?", fontsize=15)
plt.ylabel("Number of Users", fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=14)
plt.xlabel("Emotion", fontsize=15)
plt.legend(
    title="Preference",
    title_fontsize=11,
    fontsize=10,
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0
)
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table (heatmap-ready) from Q101 preferences
heatmap_data = q101_long.groupby(['Emotion', 'Preference']).size().unstack(fill_value=0)

# Normalize to get proportions (row-wise)
heatmap_prop = heatmap_data.div(heatmap_data.sum(axis=1), axis=0).round(2)

# Plot heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(heatmap_prop, annot=True, cmap='coolwarm', cbar=True, fmt=".2f")
plt.title("Proportion of Guidance vs. Freedom Preferences by Emotion (Q101)", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel("Emotion", fontsize=15)
plt.xlabel("Preference", fontsize=15)
plt.tight_layout()
plt.show()
```
